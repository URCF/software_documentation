```python
import glob
import os
import subprocess
```

```python
tmpInput = open('../urcfwiki/test/docs/software_docs/installed.md', 'r')
tmpOutput = open('installed.md','w')

listSoftware = tmpInput.readlines()
 
count = 0
# Strips the newline character
for line in listSoftware:
    count += 1
    if count <= 2:
        tmpOutput.write(line)
    else:
        fileName = line.split(':')[1].strip()
        dirName = fileName.split('.')[0].strip()
        #subprocess.run(["mkdir", dirName])
        #destFilePath = os.path.join(dirName.lower(), 'README.md')
        #srcFilePath = os.path.join('../urcfwiki/test/docs/software_docs/installed/',fileName)
        #subprocess.run(["cp", srcFilePath, destFilePath])
        tmpOutput.write("- [" + dirName + "](" + "https://github.com/URCF/software_documentation/tree/main/" + dirName.lower() + ")")
        #print(filePath)
       

tmpInput.close()
tmpOutput.close()
```
